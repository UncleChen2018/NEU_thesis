```python
import numpy as np
import matplotlib.pyplot as plt

def plot_np_array(array):
    # Ensure the input is a 1D array
    if array.ndim != 1:
        raise ValueError("Input must be a 1D NumPy array.")

    # Create the plot
    plt.figure(figsize=(10, 5))  # Set the figure size (optional)
    plt.plot(array, marker='o', linestyle='-', color='b',linewidth=1, markersize=3)  # Plot the array data
    plt.title('Plot of the 1D Array')  # Title of the plot
    plt.xlabel('Index')  # X-axis label
    plt.ylabel('Value')  # Y-axis label
    plt.grid(True)  # Show grid
    plt.show()  # Display the plot

# Example usage:
data = np.random.randn(100)  # Create a random 1D array of size 100
plot_np_array(data)
```

```python
import numpy as np
from scipy.stats import entropy

def calculate_entropy_with_scipy(data):
    # Count the frequency of each unique value
    values, counts = np.unique(data, return_counts=True)
    # Calculate the entropy using scipy.stats.entropy
    return entropy(counts, base=2)

# Example usage
np_array = np.random.randint(0, 10, size=1000)  # Smaller range for clearer example
print("Entropy:", calculate_entropy_with_scipy(np_array))
```

```python
import gzip

def elias_gamma_encode(numbers):
    def encode_number(n):
        # Map positive to even and negative to odd using 2x and 2x-1 respectively
        mapped_n = 2 * abs(n) if n >= 0 else 2 * abs(n) - 1
        # Elias gamma encode the mapped number which is always positive
        binary_rep = bin(mapped_n + 1)[2:]  # Binary representation of mapped_n + 1
        unary_len = len(binary_rep) - 1  # Unary length part (length of binary - 1)
        unary_rep = "0" * unary_len + "1"  # Unary representation
        return (
            unary_rep + binary_rep[1:]
        )  # Combine unary with binary part except the first '1'

    # Encode each number in the list and concatenate
    encoded = "".join(encode_number(num) for num in numbers)
    # Pad the encoded string to make its length a multiple of 8
    padding_length = (8 - len(encoded) % 8) % 8
    padded_encoded = encoded + "0" * padding_length
    # Convert the padded binary string to bytes
    byte_array = bytes(
        int(padded_encoded[i : i + 8], 2) for i in range(0, len(padded_encoded), 8)
    )
    return byte_array, padding_length


def elias_gamma_decode(byte_array, padding_length):
    # Convert bytes back to binary string
    encoded = "".join(f"{byte:08b}" for byte in byte_array)
    # Remove padding
    encoded = encoded[:-padding_length] if padding_length > 0 else encoded
    numbers = []
    i = 0

    while i < len(encoded):
        # Count the number of consecutive '0's to determine the length of the binary part
        length = 0
        while i < len(encoded) and encoded[i] == "0":
            length += 1
            i += 1

        i += 1  # Skip the '1' that marks the end of the unary part
        binary_part = "1" + encoded[i : i + length]
        i += length

        # Convert binary part to integer
        mapped_x = int(binary_part, 2) - 1
        x = (mapped_x + 1) // 2
        if mapped_x % 2 == 1:
            numbers.append(-x)
        else:
            numbers.append(x)

    return numbers


# Example usage
test_numbers = (np.random.normal(loc=0, scale=1, size=10000) * 2**32).astype('int64')
encoded_bytes, padding_length = elias_gamma_encode(test_numbers)
decoded = elias_gamma_decode(encoded_bytes, padding_length)

# print("Encoded Bytes:", encoded_bytes)
# print("Decoded:", decoded)

print(np.array_equal(test_numbers, decoded))
print(len(encoded_bytes), len(test_numbers) * 4)
gzip_encoded = gzip.compress(encoded_bytes)
print(len(gzip_encoded), len(test_numbers) * 4)
```

```python
import numpy as np
import io
import zstandard as zstd

def compress_data_and_print_ratio(data, mode='b', method='gzip'):
    # Check the mode and convert the numpy array accordingly
    if isinstance(data, np.ndarray):
        if mode == 'b':
            # Convert all types directly to bytes
            data = data.tobytes()
            # print('Converted to bytes')
        elif mode == 't':
            # Convert all types to a text representation and then to bytes
            data = '\n'.join(map(str, data)).encode('utf-8')
            # print('Converted to text')

    # Initialize a BytesIO object to store the compressed data
    buffer = io.BytesIO()

    if method == 'gzip':
        # Compress the data using gzip
        with gzip.GzipFile(fileobj=buffer, mode='wb',compresslevel=9) as f:
            f.write(data)
    elif method == 'zstd':
        # Compress the data using zstandard
        compressor = zstd.ZstdCompressor(level=22)
        buffer.write(compressor.compress(data))
    else:
        raise ValueError("Unsupported compression method: {}".format(method))

    # Retrieve the compressed data from the buffer
    compressed_data = buffer.getvalue()
    compressed_size = len(compressed_data)
    original_size = len(data)

    # Calculate and print the compression ratio
    compression_ratio = original_size / compressed_size
    # print("Original size (bytes):", original_size)
    # print("Compressed size (bytes):", compressed_size)
    # print("Compression ratio:", compression_ratio)

    return compressed_size

# Example usage:
# Create a numpy array with float32 data
float_data = np.array([1.1, 2.2, 3.3], dtype=np.float32)

compress_data_and_print_ratio(float_data, 'b')

# Test the function with Zstd and high compression level
compress_data_and_print_ratio(float_data, 'b', 'zstd')
```

```python
import numpy as np
import os
import subprocess

def compress_with_sz3(array, EB):
    # Write the bytes to a binary file
    temp_input_filename = 'sz_temp.bin'
    temp_output_filename = 'sz_temp.bin.sz'
    
    # Convert the numpy array to bytes and write to file
    array.tofile(temp_input_filename)
    
    # Prepare the SZ3 command based on the dtype of the array
    if array.dtype == np.float32:
        dtype_flag = '-f'
    elif array.dtype == np.float64:
        dtype_flag = '-d'
    elif array.dtype == np.int32:
        dtype_flag = '-I 32'
    elif array.dtype == np.int64:
        dtype_flag = '-I 64'
    else:
        raise ValueError("Unsupported data type for SZ3 compression")


    # Print the command for verification
    # command_string = " ".join(command)
    # 


    command = f"./sz3 {dtype_flag} -i {temp_input_filename} -z {temp_output_filename} -1 {array.size} -M ABS {EB}"
    print("Executing command:", command)
    # Execute the command
    subprocess.run(command,shell=True, check=True)
    
    # Get the size of the compressed file
    compressed_size = os.path.getsize(temp_output_filename)
    
    # Remove the temporary files
    os.remove(temp_input_filename)
    os.remove(temp_output_filename)
    
    return compressed_size

# Example usage for float32 data
data_float = np.array([1.1, 2.2, 3.3, 4.4] * 1000, dtype=np.float32)
error_bound = 0.01
compressed_file_size = compress_with_sz3(data_float, error_bound)
print("Compressed file size for float data:", compressed_file_size)

# Example usage for int32 data
data_int = np.array([1, 2, 3, 4] * 1000, dtype=np.int32)
compressed_file_size_int = compress_with_sz3(data_int, error_bound)
print("Compressed file size for int data:", compressed_file_size_int)
```
